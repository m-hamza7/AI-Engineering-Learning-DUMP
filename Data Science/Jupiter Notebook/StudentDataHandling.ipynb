```python
df["Position"]= df["percentage"].rank(ascending=False, method="dense").astype(int)

print(df)
```

```python
for st in students:
    position = df[df['Name']==st]['Position'].values[0]
    if position == 1:
         print("Position of", st, ": ", position,"st")
    elif position == 2:
        print("Position of", st, ": ", position,"nd")
    elif position == 3:
        print("Position of", st, ": ", position,"rd")
    else:
        print("Position of", st, ": ", position,"th")
        

```

```python
def rankSuffix(n):
    if 11<=n%100 <=13:
        return "th"
    return {1:"st", 2:"nd", 3:"rd"}.get(n%10, "th")


for st in students:
    position = df[df['Name']==st]['Position'].values[0]
    print(f"\n{st}: ", f"{position}{rankSuffix(position)}")
```

```python
df["Position_Rank"] = df["Position"].apply(lambda x: f"{x}{rankSuffix(x)}")

print(df[["Name", "Position_Rank"]])

```
