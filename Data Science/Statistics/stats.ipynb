```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as mt 

df = pd.read_csv('StudentData.csv')

print(df)
```

```python
avg_total_marks = df['Total'].mean()
print("Average total marks of the class are:",avg_total_marks)
```

```python
median = df['Total'].median()
print("Median of total marks is:",median)
```

```python
mode = df['Total'].mode()
print("Mode of total marks is:",mode)
```

```python
Range = df['Total'].max() - df['Total'].min()
print("Range of total marks is:",Range)
```

```python
Variance = df['Total'].var()
print("Variance of total marks is:",Variance)
```

```python
Standard_Deviation = df['Total'].std()
print("Standard Deviation of total marks is:",Standard_Deviation)
```

```python
df2 = {  'Name': ['Ali', 'Sara', 'Ahmed', 'Usman', 'Ayesha', 'Bilal', 'Zara'],
        'HoursStudied': [2, 5, 1, 3, 6, 2, 4],
        'Marks': [55, 85, 45, 65, 92, 58, 75] }
#print(df2)
df3 = pd.DataFrame(df2)
print(df3)
```

```python
mean2  = df3['Marks'].mean()
print("Mean of marks is:",mean2)
median2 = df3['Marks'].median()
print("Median of marks is:",median2)    
mode2 = df3['Marks'].mode()
print("Mode of marks is:",mode2)
Range2 = df3['Marks'].max() - df3['Marks'].min()
print("Range of marks is:",Range2)     

Variance2 = df3['Marks'].var()
print("Variance of marks is:",Variance2)

Standard_Deviation2 = df3['Marks'].std()
print("Standard Deviation of marks is:",Standard_Deviation2)

correlation = df3['HoursStudied'].corr(df3['Marks'])
print("Correlation between hours studied and marks is:",correlation)
```
