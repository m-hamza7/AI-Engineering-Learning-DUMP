```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Hours':[2,3,4,7,8,9],
    'Pass':[0,0,0,1,1,1]
})


print(df)

new_hours = 6

k=3

df['Distance'] = np.abs(df['Hours'] - new_hours)

df = df.sort_values('Distance')

print(df)

df_sorted = df.sort_values('Distance')
print(df_sorted)

neighbors = df_sorted[:k]
print(neighbors)

prediction  = neighbors['Pass'].mode()[0]


print(f'Predicted Pass value for {new_hours} hours is: {prediction}')


```

```python
df2 = pd.DataFrame({
    'Hours':[1,5,6,6,5,4],
    'Attendance':[90,45,85,90,80,70],
    'pass':[0,0,1,1,1,0]
})


new_student_hours = 5
new_student_attendance = 80
k=3

df2['Distance'] = np.sqrt((df2['Hours'] - new_student_hours)**2 + (df2['Attendance'] - new_student_attendance)**2)

df2_sorted = df2.sort_values('Distance')
print(df2_sorted)

neighbors2 = df2_sorted[:k]
print("\n",neighbors2)

predictions = neighbors2['pass'].mode()

print(f'Predicted pass value for student with {new_student_hours} hours and {new_student_attendance}% attendance is: {predictions[0]}') 
```
