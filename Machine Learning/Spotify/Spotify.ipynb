```python
import pandas as pd
import matplotlib.pyplot as pt


df_album = pd.read_csv('spotify_skz_counts_album.csv')
df_streaming = pd.read_csv('spotify_skz_streaming_4.25_4.26.csv', encoding='latin1')
df_days = pd.read_csv('spotify_skz_counts_day.csv')


print("album data shape:", df_album.shape,"\n\n")

print("album info\n\n")
df_album.info()
print("\n\n")


print("Describe\n\n")
print(df_album.describe())
print("\n\n")


print("NUll\n\n")
print(df_album.isnull().sum())
print("\n\n")



pt.hist(df_album['number_of_streams'])
pt.title('Number of Streams Histogram')
pt.xlabel('Number of Streams')  
pt.ylabel('Frequency')
pt.show()

pt.hist(df_album['bpm'])
pt.title('BPM Histogram')
pt.xlabel('BPM')  
pt.ylabel('Frequency')
pt.show()


corr = df_album.corr(numeric_only=True)
print("Correlation Matrix\n\n")
print(corr)
```
